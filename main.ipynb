```python
sine_wave_data = df[df['励磁波形'] == '正弦波'].head().iloc[0, 4:].values
```

```python
import matplotlib.pyplot as plt

# Filter the data for '正弦波' waveforms
sine_wave_data = df[df['励磁波形'] == '正弦波'].head().iloc[0, 4:].values
triangle_wave_data = df[df['励磁波形'] == '三角波'].head().iloc[0, 4:].values
trapezoidal_wave_data = df[df['励磁波形'] == '梯形波'].head().iloc[2, 4:].values

# Plot the waveforms on the same figure
plt.plot(sine_wave_data, label='正弦波')
plt.plot(triangle_wave_data, label='三角波')
plt.plot(trapezoidal_wave_data, label='梯形波')
plt.title('Waveform Plot for Different Waveforms')
plt.xlabel('Sample Points')
plt.ylabel('Magnetic Flux Density (B, T)')
plt.legend()
plt.show()
```
